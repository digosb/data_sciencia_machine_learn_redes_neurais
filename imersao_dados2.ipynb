```python
# Retirando os dados NaN e as notas zeros
dados_sem_zero = dados[provas].query('(NU_NOTA_REDACAO > 0) & (NU_NOTA_CH != 0 )').dropna()
dados_sem_zero.head()
```

```python
len(dados_sem_zero)
```

```python
# Abriando os dados conforme foi reitado acima
provas_entrada = ['NU_NOTA_CN','NU_NOTA_CH','NU_NOTA_LC','NU_NOTA_REDACAO']
provas_saida = 'NU_NOTA_MT'
```

```python
notas_entrada = dados_sem_zero[provas_entrada]
notas_saida = dados_sem_zero[provas_saida]
```

```python
notas_entrada.sample()
```

```python
notas_saida.sample()
```

```python
# Fecture
X = notas_entrada

# Labels
y = notas_saida
```

```python
X.sample()
```

```python
y.sample()
```

```python
# Importando bibliotecas
from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVR
from sklearn.dummy import DummyRegressor
from sklearn.metrics import mean_squared_error
```

```python
# Treinando modelo

SEED = 4321
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=SEED)

modelo = LinearSVR(random_state=SEED)
modelo.fit(X_train,y_train)
```

```python
# Realizando previsão

pred_mt = modelo.predict(X_test)
```

```python
pred_mt
```

```python
# Criando novo DataFrame para comparar os dados reias com os previtos. 
resultados = pd.DataFrame()
resultados['Real'] = y_test
resultados ['Previsao'] = pred_mt
resultados['diferenca'] = resultados['Real'] - resultados ['Previsao']
resultados['quadrado_diferenca'] = (resultados['Real'] - resultados ['Previsao'])**2
```

```python
resultados.head()
```

```python
resultados['quadrado_diferenca'].mean()
```

```python
resultados['quadrado_diferenca'].mean()**(1/2)
```

```python
# Treinando, prevendo e calculando erro do modelo 
modelo = LinearSVR(random_state=SEED)
modelo.fit(X_train,y_train)
pred_mt = modelo.predict(X_test)
mean_squared_error(y_test, pred_mt)
```

```python
from sklearn.tree import DecisionTreeRegressor

modelo_arvore = DecisionTreeRegressor(max_depth = 3)
modelo_arvore.fit(X_train, y_train)
pred_mt_arvore = modelo_arvore.predict(X_test)
mean_squared_error(y_test, pred_mt_arvore)
```

```python
from sklearn.model_selection import cross_validate

modelo_arvore = DecisionTreeRegressor(max_depth = 3)
resultados_cross = cross_validate(modelo_arvore, X,y, cv=10, scoring="neg_mean_squared_error")
media = (resultados_cross['test_score']*-1).mean()
desvio_padrao = (resultados_cross['test_score']*-1).std()
lim_inferiror = media - (2*desvio_padrao)
lim_superior = media + (2*desvio_padrao)

print(f"Intervalo de confiança esta entre {lim_inferiror} - {lim_superior}")
```

```python
# Criando Função
def calcula_mse(resultados):
  media = (resultados_cross['test_score']*-1).mean()
  desvio_padrao = (resultados_cross['test_score']*-1).std()
  lim_inferiror = media - (2*desvio_padrao)
  lim_superior = media + (2*desvio_padrao)
  print(f"Intervalo de confiança esta entre {lim_inferiror} - {lim_superior}")
```

```python
from sklearn.model_selection import cross_validate
from sklearn.model_selection import KFold
import numpy as np

SEED = 1234
np.random.seed(SEED)


partes = KFold(n_splits = 10, shuffle=True) #dividindo os dados em partes 
modelo_arvore = DecisionTreeRegressor(max_depth = 3)
resultados_cross = cross_validate(modelo_arvore, X,y, cv=partes, scoring="neg_mean_squared_error")
calcula_mse(resultados)
```

```python
def regressao_arvore(nivel):
  SEED = 1232
  np.random.seed(SEED)
  partes = KFold(n_splits = 10, shuffle=True) #dividindo os dados em partes 
  modelo_arvore = DecisionTreeRegressor(max_depth = nivel)
  resultados_cross = cross_validate(modelo_arvore, X,y, cv=partes, scoring="neg_mean_squared_error", return_train_score=True)
  print(f"Treino = {(resultados_cross['train_score']*-1).mean()} | Teste = {(resultados_cross['test_score']*-1).mean()}")

regressao_arvore(4)
```

```python
for i in range(1,21):
  regressao_arvore(i)
```
